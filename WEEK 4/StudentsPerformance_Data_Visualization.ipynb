````python
# Fungsi untuk plotting pie chart dengan warna estetik
def plot_pie_chart(column_name):

    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Warna estetik menggunakan palet "Pastel1" dari matplotlib
    colors = plt.get_cmap('Pastel1')(range(len(labels)))

    # Membuat sedikit efek explode pada setiap slice
    explode = [0.05] * len(labels)  # Membuat setiap slice terpisah sedikit

    plt.figure(figsize=(7,7))
    plt.pie(
        sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,  # Menggunakan warna dari palet Pastel1
        explode=explode,  # Menambahkan efek explode
        shadow=True,  # Menambahkan efek bayangan
        wedgeprops={'edgecolor': 'black'}  # Menambahkan garis tepi hitam untuk memperjelas batas
    )
    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}', fontsize=14, color='darkblue')  # Judul dengan font lebih besar dan warna
    plt.show()

# Menampilkan pie chart untuk kolom bertipe kategorikal
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
````

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
# Menambahkan kolom untuk status lulus atau gagal
df['math_status'] = ['Pass' if x >= 75 else 'Fail' for x in df['math score']]
df['reading_status'] = ['Pass' if x >= 75 else 'Fail' for x in df['reading score']]
df['writing_status'] = ['Pass' if x >= 75 else 'Fail' for x in df['writing score']]

# Membuat box plot untuk setiap pelajaran
def plot_boxplot():
    plt.figure(figsize=(15, 5))

    # Box plot untuk Math
    plt.subplot(1, 3, 1)
    sns.boxplot(x='math_status', y='math score', data=df, palette='pastel')
    plt.title('Math Score Distribution')
    plt.xlabel('Status')
    plt.ylabel('Math Score')

    # Box plot untuk Reading
    plt.subplot(1, 3, 2)
    sns.boxplot(x='reading_status', y='reading score', data=df, palette='pastel')
    plt.title('Reading Score Distribution')
    plt.xlabel('Status')
    plt.ylabel('Reading Score')

    # Box plot untuk Writing
    plt.subplot(1, 3, 3)
    sns.boxplot(x='writing_status', y='writing score', data=df, palette='pastel')
    plt.title('Writing Score Distribution')
    plt.xlabel('Status')
    plt.ylabel('Writing Score')

    plt.tight_layout()
    plt.show()

# Memanggil fungsi untuk menampilkan box plot
plot_boxplot()
````

Prompt:

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
# Menambahkan kolom untuk status lulus atau gagal
df['math_status'] = ['Pass' if x >= 75 else 'Fail' for x in df['math score']]
df['reading_status'] = ['Pass' if x >= 75 else 'Fail' for x in df['reading score']]
df['writing_status'] = ['Pass' if x >= 75 else 'Fail' for x in df['writing score']]

# Menghitung jumlah siswa yang pass dan fail untuk setiap pelajaran
def calculate_pass_fail_counts():
    results = {
        'Subject': ['Math', 'Reading', 'Writing'],
        'Pass': [
            df['math_status'].value_counts().get('Pass', 0),
            df['reading_status'].value_counts().get('Pass', 0),
            df['writing_status'].value_counts().get('Pass', 0)
        ],
        'Fail': [
            df['math_status'].value_counts().get('Fail', 0),
            df['reading_status'].value_counts().get('Fail', 0),
            df['writing_status'].value_counts().get('Fail', 0)
        ]
    }

    return pd.DataFrame(results)

# Mendapatkan hasil pass dan fail
pass_fail_df = calculate_pass_fail_counts()

# Membuat bar chart
def plot_pass_fail_bar_chart(df):
    df.set_index('Subject').plot(kind='bar', stacked=False, color=['lightgreen', 'salmon'])
    plt.title('Jumlah Siswa yang Lulus (Pass) dan Gagal (Fail)')
    plt.xlabel('Mata Pelajaran')
    plt.ylabel('Jumlah Siswa')
    plt.xticks(rotation=0)
    plt.legend(title='Status', labels=['Pass', 'Fail'])
    plt.tight_layout()
    plt.show()

# Memanggil fungsi untuk menampilkan grafik
plot_pass_fail_bar_chart(pass_fail_df)
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
# Menambahkan kolom nilai rata-rata dari Math, Reading, dan Writing
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Melihat nilai rata-rata berdasarkan kategori tertentu
def plot_average_score_by_category(category):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=category, y='average_score', data=df, palette='pastel')
    plt.title(f'Average Score by {category}')
    plt.xlabel(category)
    plt.ylabel('Average Score')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Melihat korelasi nilai rata-rata dengan kategori 'gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'
categories = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Plot untuk setiap kategori
for category in categories:
    plot_average_score_by_category(category)
````
